## Part 1

```python
data = pd.read_table('input-02.txt', names=['O', 'P'], sep=' ')
data
```

```python
d = (
    data
    .assign(
        iO = lambda x: x.O.map(ord) - ord('A'),
        iP = lambda x: x.P.map(ord) - ord('X'),
        delta_mod3 = lambda x: x.iP.sub(x.iO).mod(3),
        score_chosen = lambda x: x.iP.add(1),
        score_outcome = lambda x: x.delta_mod3.map({1: 6, 2: 0, 0: 3})
    )
)
d
```

```python
d[['score_chosen', 'score_outcome']].sum().sum()
```

## Part 2

```python
data = pd.read_table('input-02.txt', names=['O', 'need'], sep=' ')
data
```

```python
d = (
    data
    .assign(
        iO = lambda x: x.O.map(ord) - ord('A'),
        need_delta_mod3 = lambda x: x.need.map({'X': 2, 'Y': 0, 'Z': 1}),
        iP = lambda x: x.need_delta_mod3.add(d.iO).mod(3),
        score_chosen = lambda x: x.iP.add(1),
        score_outcome = lambda x: x.need_delta_mod3.map({1: 6, 2: 0, 0: 3})
    )
)
d
```

```python
d[['score_chosen', 'score_outcome']].sum().sum()
```
