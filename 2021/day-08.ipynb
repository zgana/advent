```python
normal_wiring = pd.Series([
    'abcefg',
    'cf',
    'acdeg',
    'acdfg',
    'bcdf',
    'abdfg',
    'abdefg',
    'acf',
    'abcdefg',
    'abcdfg'
])
```

```python
digit_possibilities = (
    normal_wiring
    .map(len)
    .reset_index()
    .set_axis('digit num_segments'.split(),1)
    .groupby('num_segments')
    .digit.agg(list)
)
digit_possibilities
```

```python
I.T[0]
```

```python
x = I.T[1].rename('uniques').to_frame()
x['uniqueset'] = x.uniques.map(set)
x['uniquetuple'] = x.uniques.map(sorted).map(tuple)
x['poss'] = digit_possibilities[x.uniques.map(len)].values
x['known'] = np.where(x.poss.map(len).eq(1), x.poss.str[0], np.nan)
x['found'] = x.known.copy()
xx = x.set_index('known')
xx
```

```python
# find 0,6,9
i069 = xx.poss.str[0].eq(0)
i6 = i069 & (xx.uniqueset - xx.loc[1].uniqueset).map(len).eq(5)
i9 = i069 & (xx.uniqueset - xx.loc[4].uniqueset).map(len).eq(2)
i0 = i069 & ~i6 & ~i9
xx.loc[i0, 'found'] = 0
xx.loc[i6, 'found'] = 6
xx.loc[i9, 'found'] = 9

# find 2,3,5
i235 = xx.poss.str[0].eq(2)
i2 = i235 & (xx.uniqueset - xx.loc[4].uniqueset).map(len).eq(3)
i3 = i235 & (xx.uniqueset - xx.loc[1].uniqueset).map(len).eq(3)
i5 = i235 & ~i2 & ~i3
xx.loc[i2, 'found'] = 2
xx.loc[i3, 'found'] = 3
xx.loc[i5, 'found'] = 5

xx
```

```python
def handle_row(i, o):
    x = i.rename('uniques').to_frame()
    x['uniqueset'] = x.uniques.map(set)
    x['uniquetuple'] = x.uniques.map(sorted).map(tuple)
    x['poss'] = digit_possibilities[x.uniques.map(len)].values
    x['known'] = np.where(x.poss.map(len).eq(1), x.poss.str[0], np.nan)
    x['found'] = x.known.copy()
    xx = x.set_index('known')
    
    # find 0,6,9
    i069 = xx.poss.str[0].eq(0)
    i6 = i069 & (xx.uniqueset - xx.loc[1].uniqueset).map(len).eq(5)
    i9 = i069 & (xx.uniqueset - xx.loc[4].uniqueset).map(len).eq(2)
    i0 = i069 & ~i6 & ~i9
    xx.loc[i0, 'found'] = 0
    xx.loc[i6, 'found'] = 6
    xx.loc[i9, 'found'] = 9

    # find 2,3,5
    i235 = xx.poss.str[0].eq(2)
    i2 = i235 & (xx.uniqueset - xx.loc[4].uniqueset).map(len).eq(3)
    i3 = i235 & (xx.uniqueset - xx.loc[1].uniqueset).map(len).eq(3)
    i5 = i235 & ~i2 & ~i3
    xx.loc[i2, 'found'] = 2
    xx.loc[i3, 'found'] = 3
    xx.loc[i5, 'found'] = 5

    assert xx.found.nunique() == 10
    
    found = xx.set_index('uniquetuple').found
    output = int(''.join(o.map(sorted).map(tuple).map(found).map(int).map(str)))
    return output
```

```python
outputs = []
for ((_,i), (_,o)) in zip(I.iterrows(), O.iterrows()):
    outputs.append(handle_row(i,o))
```
