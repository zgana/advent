## Part 1

```python
inp = pd.read_table('input-02.txt', names='dir N'.split(), sep=' ')
inp
```

```python
dxmap = dict(forward=1, up=0, down=0)
dzmap = dict(forward=0, up=-1, down=1)

inp['dx'] = inp.dir.map(dxmap) * inp.N
inp['dz'] = inp.dir.map(dzmap) * inp.N
inp
```

```python
inp['dx dz'.split()].sum().prod()
```

## Part 2

```python
inp = pd.read_table('input-02.txt', names='dir N'.split(), sep=' ')
```

```python
d_aim_map = dict(forward=0, down=1, up=-1)
dx_map = dz_map = dict(forward=1, down=0, up=0)

inp['d_aim'] = inp.dir.map(d_aim_map) * inp.N
inp['aim'] = inp.d_aim.cumsum()
inp['dx'] = inp.dir.map(dx_map) * inp.N
inp['dz'] = inp.dir.map(dz_map) * inp.N * inp.aim
inp
```

```python
inp['dx dz'.split()].sum().prod()
```
